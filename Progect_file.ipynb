```python

#student management system


#intilize dit

student_grades={"student1":30,"student2":100,"student3":00,"Ajith":850,"student5":100,"student6":160}


#add a new student
def add_student(name,grade):
    student_grades[name]=grade

    print("Added "+name+" with a "+str(grade))

#update student

def update_student(name,grade):
    if name in student_grades:
        student_grades[name]=grades

        print(name+" with marks are updated"+str(grade))
    else:
        print(name+" is not found")

#delete a student
def delete_student(name):
    if name in student_grades:
        del student_grades[name]
        print(name+" has been successfully deleted")
    else:
        print(name+" not found")

#view all students

def display_all_students():
    if student_grades:
        for name,grade in student_grades.items():
            print(name+":"+str(grade))
    else:
        print("no student found/added")



def find_grades():
# Counters
  first_class = []
  second_class = []
  third_class = []
  failed = []

  # Classify students
  for name, marks in student_grades.items():
      if marks >= 60:
          first_class.append(name)
      elif marks >= 50:
          second_class.append(name)
      elif marks >= 35:
          third_class.append(name)
      else:
          failed.append(name)

  # Results
  print(" First Class Students:", first_class)
  print(" Second Class Students:", second_class)
  print(" Third Class Students:", third_class)
  print(" Failed Students:", failed)

  print("\nSummary:")
  print("First Class:", len(first_class))
  print("Second Class:", len(second_class))
  print("Third Class:", len(third_class))
  print("Failed:", len(failed))


def first_name():
    maximum_number = 0
    pos = 0

    items = list(student_grades.items())  # convert to list of (key,value) pairs

    for i, (name, grade) in enumerate(items):
        if grade > maximum_number:
            maximum_number = grade
            pos = i   # store index

    print("Maximum grade:", maximum_number)
    print("Class Topper is:", items[pos][0])  # student name
    print("Total marks:", items[pos][1])  # grade








def main():
        print("\n student Grades Management system")
        print("1.Add student")
        print("2. update student")
        print("3.Delete student")
        print("4. View student")
        print("5. Grades student")
        print("6. Tooper of the class")
        print("7.Exit")
        while True:
          choice = int(input("Enter your choice = "))
          if choice == 1:
              name = input("Enter student name = ")
              grade = int(input("Enter student grade = "))
              add_student(name,grade)
          elif(choice==2):
              name = input("Enter student name = ")
              grade = int(input("Enter student grade = "))
              update_student(name,grade)
          elif(choice==3):
              name = input("Enter student name = ")
              delete_student(name)
          elif (choice==4):
              display_all_students()
          elif(choice==5):
             find_grades()
          elif(choice==6):
              first_name()
          elif (choice==7):
              print("closing the program")
              break
          else:
              print("Invalid choice")

main()
```
